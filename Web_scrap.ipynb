```python
#passing empty list in which I need to append the fetched data
import pandas as pd
df = pd.DataFrame()
ljob_title = []
lname_location = []
lcompany_name = []
lwork_format = []
ljob_type = []
ladd_info = []
```

```python
for j in range(1,8):#8 page empty thats why running the loop till 7
    page_numbers =  driver.find_element(By.XPATH,f"//button[@aria-label='Page {j}']")
    page_numbers.click()
    time.sleep(3)
    
    jobs_block = driver.find_elements(By.XPATH,"//ul[@class='scaffold-layout__list-container']/li")
    #executes javascript to scroll the job div using views
    for job in jobs_block:
        driver.execute_script('arguments[0].scrollIntoView();',job)
        time.sleep(1)
    #looping and scraping job data from page j
    job_links = driver.find_elements(By.XPATH,"//ul[@class = 'scaffold-layout__list-container']/li/div/div/div[1]/div[2]/div[1]/a")
    for link in job_links:
        link.click()
        time.sleep(3)
        job_title = driver.find_element(By.CSS_SELECTOR,"span[class*='top-card__job-title-link']").text
        company_name = driver.find_element(By.CSS_SELECTOR,"div[class*='primary-description-container'] a[href*='company']").text
        work_format =  driver.find_element(By.CSS_SELECTOR,'ul > li[class*="job-insight"] > span > span').text 
        job_location = driver.find_element(By.CSS_SELECTOR,"div[class*='primary-description-container'] > div").text
        job_location_text = re.search(r'Â·\s*([^Â·]+)', job_location) #to eliminate some special characters
        job_type = driver.find_element(By.CSS_SELECTOR,'ul > li[class*="job-insight"] > span > span + span').text
        add_info = driver.find_element(By.CSS_SELECTOR,'ul > li[class*="job-details"] + li  > span').text
        ljob_title.append(job_title)
        lname_location.append(job_location)
        lcompany_name.append(company_name)
        lwork_format.append(work_format)
        ljob_type.append(job_type)
        ladd_info.append(add_info)
```

```python
data = {'job_title':ljob_title,'company_name':lcompany_name,'work_format':lwork_format,'job_type':ljob_type,'loc_name':lname_location,'Additional_info': ladd_info}
df = pd.DataFrame(data)
```
